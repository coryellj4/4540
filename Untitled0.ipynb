First we will import numpy and matplot for some datastrucutre and other needed functions.
We will also go ahead and load in our data as the sets x and y.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 4, 7, 10])
y = np.array([1.5, 3.5, 9, 8])
m = len(x)
```

Next we will go ahead and define a predict function, which simply caculates a prediction based on our current theta values, and a cost function based on squared distance which we will use later

```python
def predict(x, theta0, theta1):
    return theta0 + theta1 * x

# Cost Function (MSE)
def cost_function(theta0, theta1):
    predictions = predict(x, theta0, theta1)
    return (1 / (2 * m)) * np.sum((y - predictions)**2)
```

Next we have the gradient descent function, which will calculate the predictions and the cost each iteration and then adjust the parameters acordingly.

```python
def gradient_descent(alpha, iterations):
    theta0 = 0
    theta1 = 0
    cost_history = []

    for i in range(iterations):
        predictions = predict(x, theta0, theta1)
        error = predictions - y

        d_theta0 = (1 / m) * np.sum(error)
        d_theta1 = (1 / m) * np.sum(error * x)

        theta0 = theta0 - alpha * d_theta0
        theta1 = theta1 - alpha * d_theta1

        cost = cost_function(theta0, theta1)
        cost_history.append(cost)

        if i % 100 == 0:
            print(f"Iteration {i}: theta0 = {theta0:.4f}, theta1 = {theta1:.4f}, Cost = {cost:.4f}")

    return theta0, theta1, cost_history
```

Next we will set the learning rate (which should be very small) and number of iterations (usally quite large)and call the gradient descent.

```python
alpha = 0.01  # Learning rate
iterations = 1000  # Number of iterations

# Run Gradient Descent
theta0, theta1, cost_history = gradient_descent(alpha, iterations)

print("\nOptimal Parameters:")
print(f"theta0 = {theta0:.4f}")
print(f"theta1 = {theta1:.4f}")
```

We can also generate some graphs with this data to beterunderstand the process, the oen on the left shows the final line after 1000 iterations as we can see it seems to fit the graph quite well. The one on the right show the cost history during the training process, and how the model gets more efficent over time

```python
plt.figure(figsize=(12, 6))

# Plot the data and the regression line
plt.subplot(1, 2, 1)
plt.scatter(x, y, label="Data Points")
x_range = np.linspace(x.min() - 1, x.max() + 1, 100)  # For smooth line
plt.plot(x_range, predict(x_range, theta0, theta1), color='red', label="Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)

# Plot the cost function history
plt.subplot(1, 2, 2)
plt.plot(cost_history)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Function History")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Finally we can use this line to make an actual prediction based on data not in the training set.

```python
new_x = 5
predicted_y = predict(new_x, theta0, theta1)
print(f"\nPrediction for x = {new_x}: y = {predicted_y:.4f}")
```
