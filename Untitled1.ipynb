First we will import numpy for some math functions,nd matplot for some graph visulization later

```python
import numpy as np
import matplotlib.pyplot as plt
```

Next we will load in our data points as two arrays, the first is the xy cooridinats and the second is the labels. You can change this or add more points. We add a third colum of all ones to the data, this is the bias term. We also load in some initail values for the weights.

```python
X = np.array([[-2, 1], [-1, 3], [-1, -1], [2, -2], [3, -1], [2, 1]])
y = np.array([0, 1, 0, 1, 1, 0])  # Labels (0 or 1)

X_bias = np.hstack((np.ones((X.shape[0], 1)), X))

theta = np.array([-0.5, 0, 7]) # Initial values for theta (bias, theta1, theta2)
```

Next we need a sigmoid function to squash the values of our output. We also have a cost function to grade our predictions.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, theta):
    m = len(y)  # Number of training examples
    z = np.dot(X, theta)  # Linear combination
    h = sigmoid(z)  # Sigmoid function
    # Logistic regression cost function
    cost = -1/m * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost
```

Finally we need a gradient descent function to adjust the weights.

```python
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)  # Number of training examples
    cost_history = []

    for i in range(iterations):
        z = np.dot(X, theta)
        h = sigmoid(z)
        gradient = (1/m) * np.dot(X.T, (h - y))
        # Update theta
        theta -= learning_rate * gradient
        cost_history.append(compute_cost(X, y, theta))

    return theta, cost_history
```

Now we can run gradient descent and see how the parameters are changed.

```python
learning_rate = 0.1
iterations = 1000
theta_optimal, cost_history = gradient_descent(X_bias, y, theta, learning_rate, iterations)

# Final learned parameters
print(f"Optimal parameters: {theta_optimal}")
```

Finally herese some graphs showing how the cost decresases and converges during training.

```python
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Convergence")
plt.show()

# 10. Predict the outcome for a new data point
new_data_point = np.array([1, 0, 2])  # x0 = 1, x1 = 0, x2 = 2
prediction = sigmoid(np.dot(new_data_point, theta_optimal))
print(f"Prediction for new data point [1, 0, 2]: {prediction:.4f}")
```

Here is another graph showing the points and the new line that we trained to seperate them

```python
plt.figure(figsize=(8, 6))

# Plot the points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k', s=100)

# Plot decision boundary
x1_range = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)
x2_range = -(theta_optimal[0] + theta_optimal[1] * x1_range) / theta_optimal[2]

# Decision boundary line
plt.plot(x1_range, x2_range, label="Decision Boundary", color='black', lw=2)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Logistic Regression Decision Boundary")
plt.legend()
plt.grid(True)
plt.show()
```
