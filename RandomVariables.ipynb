```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli, uniform, poisson, norm, lognorm

# Setting the style for plots
sns.set(style="whitegrid")
```

# Bernoulli Distribution

Equation for Bernoulli Distribution:<br>
P(X=x)= p if x=1 (success) <br>
&emsp;&emsp;&emsp;  = 1-p if x=0 (failure)<br>
Where p is the probabilty of success.





```python
p = 0.3  # Probability of success
n_samples = 1000

# Generating samples
bernoulli_samples = bernoulli.rvs(p, size=n_samples)

# Plotting Histogram and PDF
plt.figure(figsize=(10, 6))
plt.hist(bernoulli_samples, bins=2, density=True, alpha=0.6, color='b', label='Histogram')

# Plotting theoretical PDF
x = np.array([0, 1])
y = [1 - p, p]
plt.stem(x, y, 'r', markerfmt='ro', label='PDF', basefmt=" ")

plt.title('Bernoulli Distribution - p = {}'.format(p))
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

# Uniform Distribution

Equation for Uniform Distribution:<br>
f(x) = 1/b-a for a≤x≤b, <br>
Where a and b are the lower and upper bounds respectively

```python
a, b = 0, 10  # Bounds
uniform_samples = uniform.rvs(loc=a, scale=b-a, size=n_samples)

# Plotting Histogram and PDF
plt.figure(figsize=(10, 6))
plt.hist(uniform_samples, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Plotting theoretical PDF
x = np.linspace(a, b, 1000)
y = uniform.pdf(x, loc=a, scale=b-a)
plt.plot(x, y, 'r-', label='PDF')

plt.title('Uniform Distribution - a = {}, b = {}'.format(a, b))
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# Poisson  Distribution

Equation for Poisson  Distribution:<br>
P(X=k)= =(λ^ke^(−λ))​/k!  ,k=0,1,2,…
&emsp;&emsp;&emsp;  <br>
Where λ is the mean rate of events occuring in a fixed interval.

```python
lambda_poisson = 5  # Mean of Poisson distribution
poisson_samples = poisson.rvs(mu=lambda_poisson, size=n_samples)

# Plotting Histogram and PDF
plt.figure(figsize=(10, 6))
plt.hist(poisson_samples, bins=30, density=True, alpha=0.6, color='c', label='Histogram')

# Plotting theoretical PDF (Poisson is discrete, so we'll use a stem plot)
x = np.arange(0, np.max(poisson_samples) + 1)
y = poisson.pmf(x, mu=lambda_poisson)
plt.stem(x, y, 'r', markerfmt='ro', label='PDF', basefmt=" ")

plt.title('Poisson Distribution - λ = {}'.format(lambda_poisson))
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

#Normal Distribution

Equation for Gaussian Normal:<br>
image.png<br>
Where μ is the mean, and σ2 is the variance (standard deviation is σ).

```python
# Gaussian Distribution
mu, sigma = 0, 1  # Mean and standard deviation
gaussian_samples = norm.rvs(loc=mu, scale=sigma, size=n_samples)

# Plotting Histogram and PDF
plt.figure(figsize=(10, 6))
plt.hist(gaussian_samples, bins=30, density=True, alpha=0.6, color='m', label='Histogram')

# Plotting theoretical PDF
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, y, 'r-', label='PDF')

plt.title('Gaussian Distribution - μ = {}, σ = {}'.format(mu, sigma))
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# Log-Normal Distribution

Equation for Log-Normal Distribution:<br>
LogNormal.PNG
<br>Where μμ and σσ are the parameters of the underlying normal distribution.

```python
# Log-Normal Distribution
mu_lognorm, sigma_lognorm = 0, 1  # Mean and standard deviation of the underlying normal distribution
lognorm_samples = lognorm.rvs(sigma_lognorm, loc=mu_lognorm, scale=np.exp(mu_lognorm), size=n_samples)

# Plotting Histogram and PDF
plt.figure(figsize=(10, 6))
plt.hist(lognorm_samples, bins=30, density=True, alpha=0.6, color='y', label='Histogram')

# Plotting theoretical PDF
x = np.linspace(0, 10, 1000)
y = lognorm.pdf(x, sigma_lognorm, loc=mu_lognorm, scale=np.exp(mu_lognorm))
plt.plot(x, y, 'r-', label='PDF')

plt.title('Log-Normal Distribution - μ = {}, σ = {}'.format(mu_lognorm, sigma_lognorm))
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.show()
```
